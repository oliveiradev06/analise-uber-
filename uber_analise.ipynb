```python
import matplotlib.pyplot as plt
```

```python
bins = [0, 2, 5, 10, 15, 20, 30]  # você escolhe os intervalos
labels = ["0-2km", "2-5km", "5-10km", "10-15km", "15-20km", "20-30km"]

df['Faixa Distância'] = pd.cut(df['Distancia da Corrida'], bins=bins, labels=labels)

plt.figure(figsize=(10,6))
df['Faixa Distância'].value_counts().sort_index().plot(kind='bar')
plt.title("Frequência por Faixa de Distância")
plt.xlabel("Faixa de Distância")
plt.ylabel("Número de Corridas")
plt.show()
```
